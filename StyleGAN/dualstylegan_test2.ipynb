```python
#@title Run

import random
import torchvision
from IPython.display import clear_output

mode = 'cartoon'  #@param ['cartoon', 'caricature', 'anime', 'arcane', 'comic', 'pixar', 'slamdunk']


# Load image and transform
#feature = transform(Image.open(img_name)).unsqueeze(dim=0).to('cuda')
feature = transform(run_face_alignment(img_name)).unsqueeze(dim=0).to('cuda')

# Download checkpoints
download_ckpt_by_mode(mode)

# Load weight
ckpt = torch.load(f'checkpoint/{mode}/generator.pt', map_location=lambda storage, loc: storage)
generator.load_state_dict(ckpt["g_ema"])
generator = generator.to('cuda')

# Load local style image and transform
style = transform(Image.open(img_name_style)).unsqueeze(dim=0).to('cuda')

# Encode local style image and get external style code
_, exstyle = encoder(style, randomize_noise=False, return_latents=True, z_plus_latent=True, return_z_plus_latent=True, resize=False)
exstyle = exstyle.detach()

#update
# Load cartoon style code from the model
# You can change the index to get different cartoon styles
cartoon_style = generator.generator.style(torch.randn(1, 512).to('cuda'))
cartoon_style = cartoon_style.detach()

# Interpolate between the external style code and the cartoon style code
# You can change the alpha value to control the degree of interpolation
alpha = 0.9
exstyle = alpha * exstyle + (1 - alpha) * cartoon_style

with torch.no_grad():
  torch.cuda.ipc_collect()
  torch.cuda.empty_cache()
  img_rec, instyle = encoder(
    feature,
    randomize_noise=False,
    return_latents=True,
    z_plus_latent=True,
    return_z_plus_latent=True,
    resize=False
  )
  img_rec = torch.clamp(img_rec.detach(), -1, 1)
  latent = exstyle.repeat(2, 1, 1)
  latent[1, 7:18] = instyle[0, 7:18]


  #update
  # Generate image with local style code
  img_gen, _ = generator([instyle.repeat(2, 1, 1)], latent, z_plus_latent=True, truncation=0.7, truncation_latent=0, use_res=True, interp_weights=[0.7]*7 + [1]*11)
  img_gen = torch.clamp(img_gen.detach(), -1, 1)

  # Generate image with local style code and original color
  img_gen2, _ = generator([instyle], latent[:1], z_plus_latent=True, truncation=0.7, truncation_latent=0, use_res=True, interp_weights=[0.7]*7 + [0]*11)
  img_gen2 = torch.clamp(img_gen2.detach(), -1, 1)

#update
# Convert the content image and the style image to tensors
content = transform(Image.open(img_name)).unsqueeze(dim=0).to('cuda')
style = transform(Image.open(img_name_style)).unsqueeze(dim=0).to('cuda')

# Resize the content image and the style image to match the other images
content = F.adaptive_avg_pool2d(content, 256)
style = F.adaptive_avg_pool2d(style, 256)
img_rec = F.adaptive_avg_pool2d(img_rec, 256)
img_gen = F.adaptive_avg_pool2d(img_gen, 256)
img_gen2 = F.adaptive_avg_pool2d(img_gen2, 256)

# Concatenate all the images together
images = torch.cat([content, img_rec,style,img_gen, img_gen2], dim=0)

# Make a grid of images
vis = torchvision.utils.make_grid(images, 5, 1)

merged = Image.fromarray(tensor_to_arr(vis))
img_rec = Image.fromarray(tensor_to_arr(img_rec))
img_gen = Image.fromarray(tensor_to_arr(img_gen))
img_gen2 = Image.fromarray(tensor_to_arr(img_gen2))

clear_output()
display(merged)
```
