# 图片生成
已经训练好各个风格的模型了，现在给定测试图片，去生成各个风格的风格图

```python
from main import build_parser
from train import network_train
from test import network_test
import os
from pathlib import Path

# 当前目录
#current_directory = os.path.dirname(os.path.abspath(__file__))
current_directory = os.getcwd()
```

现在执行生成，每张图片，每个风格都要执行一次

```python
weight_parentdir = "trained_models"
weightparam_subdir = "StyleImages10_300_3_10000"


# 内容图目录
contentimage_dir = os.path.join(current_directory, "..", "Images", "original")
# 所有内容图文件名
content_images = os.listdir(contentimage_dir)

weight_parentdir = os.path.join(current_directory, "trained_models", weightparam_subdir)
styleweight_dirs = os.listdir(weight_parentdir)

# 输出目录
output_dir = os.path.join(current_directory, "output")
for styleweight_dir in styleweight_dirs:
    for image_filename in content_images:
        split_tup = os.path.splitext(image_filename)
        filename_prefix = split_tup[0]
        filename_suffix = split_tup[1]
        test_content = os.path.join(contentimage_dir, image_filename)

        
        model_load_path = os.path.join(weight_parentdir, styleweight_dir,"transform_network.pth")
        curfile_outputdir = os.path.join(output_dir, weightparam_subdir, filename_prefix)
        if not os.path.exists(curfile_outputdir):
            os.makedirs(curfile_outputdir)
        
        output_path = os.path.join(curfile_outputdir, filename_prefix + "_" + styleweight_dir + filename_suffix)
        custom_args = [
                "--train-flag", "False",
                "--model-load-path", model_load_path,
                "--test-content", test_content,
                "--output", output_path,        
            ]

        parser = build_parser()
        args = parser.parse_args(custom_args)
        network_test(args)
```
